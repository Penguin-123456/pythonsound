```python
# @title 初期設定
import numpy as np #numpyのインストール
import IPython #IPythonのインストール
```

```python
# @title 音を鳴らす
duration = 5.0 # @param{type: "number"}
rate = 48000
hz = 9900000 # @param {type:"slider", min:1, max:9900000, step:1}
t = np.linspace(0., duration, int(rate*duration))
x = np.sin(2.0*np.pi*40000*t)
IPython.display.Audio(x, rate=rate, autoplay=True)
```

# **使い方**
1.   初期設定の場所を▶を押すか`Ctrl + Enter`を押してください・
2.   durationのところは秒数です。好きな秒数を選んでください。
3.   下のスライダー(hz)を動かしてください。
4.   次に`Ctrl + Enter`を押すか、▶を押してください。
5.   自動で再生されますが、再生できない場合は▶を押してください。
